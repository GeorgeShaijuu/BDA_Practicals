```python
import matplotlib.pyplot as plt
import numpy as np

#SAMPLE DATA
students = ["Tom", "Bob", "Rahul", "Adit", "Mohith"]
math_score = [80, 82, 77, 89, 90]
sci_score = [82, 92, 89, 87, 85]
eng_score = [88, 92, 89, 99, 100]

plt.figure(figsize=(10, 8))
x = np.arange(len(students)) # Use numpy.arange to create positions for the bars

plt.bar(x, math_score, width=0.2, label="Math", align ='center')
plt.bar(x+0.2, sci_score, width=0.2, label="Science", align ='center')
plt.bar(x+0.4, eng_score, width=0.2, label="English", align ='center')

plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores by Subject")
plt.xticks(x + 0.2, students) # Set x-axis ticks to student names
plt.legend()
plt.show()
```
